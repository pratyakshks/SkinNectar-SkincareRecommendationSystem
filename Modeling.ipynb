```python
concern_solution = {'acne':'salicylic acid','dryness':'hyaluronic acid',
                    'hyperpigmentation':'vitamin c','wrinkles':'retinol','dark spots':'niancinamide',
                    'uneven texture':'glycolic acid','dullness':'vitamin C + E acid'}
```

```python
option_1
```

```python
Skin_type = {1:'Combination',2:'Dry',3:'Normal',4:'Oily',5:'Sensitive'}
```

```python
type = int(input("Enter your skin type (Press 1 for Combination/2 for Dry/3 for Normal/4 for Oily/ 5 for Sensitive) : "))
concern1 = input("Enter your Primary skin concern(Acne/Dryness/Hyperpigmentation/Wrinkles/Dark Spots/Uneven Texture/Dullness) :")
concern2 = input("Enter your Primary skin concern(Acne/Dryness/Hyperpigmentation/Wrinkles/Dark Spots/Uneven Texture/Dullness) :")
concern1 = concern1.lower()
concern2 = concern2.lower()
price = int(input("Enter your Price range (in dollars) : "))

solution1 = concern_solution[concern1]
solution2 = concern_solution[concern2]
type = Skin_type[type]

options = data[data[type] == 1][data['price']<=price]
options = options.sort_values(by = 'rank', ascending = False)
options = options[['Label','brand','name','URL','price','rank','ingredients']]

options = options[options['ingredients'].str.contains(solution1 or solution2, case=False)]
```

```python
opt_cle_1 = options[options['Label'] == 'Cleanser']
opt_mask_2 = options[options['Label'] == 'Face Mask']
opt_tre_3 = options[options['Label'] == 'Treatment']
opt_eye_4 = options[options['Label'] == 'Eye cream']
opt_mois_5 = options[options['Label'] == 'Moisturizer']
opt_sun_6 = options[options['Label'] == 'Sun protect']
```

```python
print(f"Your Skin Type - {type}\nPrice Range - {price}\nPrimary Skin Concern - {concern1}\nSecondary Skin Concern - {concern2}")
```
