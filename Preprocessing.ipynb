MODELLING

```python
import pandas as pd
import numpy as np 
from sklearn.manifold import TSNE
```

```python
data_pre = pd.read_csv("cosmetic_p.csv")
```

```python
option_1 = data_pre.Label.unique().tolist()
option_2 = data_pre.columns[6:].tolist()
```

```python
option_1
```

```python
option_2
```

```python
def my_recommender(op_1, op_2):
    df = data_pre[data_pre['Label'] == op_1][data_pre[op_2] == 1]
    df = df.reset_index()

    # embedding each ingredients
    ingredient_idx = {}
    corpus = []
    idx = 0

    for i in range(len(df)):
        ingreds = df['ingredients'][i]
        ingreds = ingreds.lower()
        tokens = ingreds.split(', ')
        corpus.append(tokens)
        for ingredient in tokens:
            if ingredient not in ingredient_idx:
                ingredient_idx[ingredient] = idx
                idx += 1

    M = len(df)                 
    N = len(ingredient_idx)     

    A = np.zeros(shape = (M, N))

    def oh_encoder(tokens):
        x = np.zeros(N)
        for t in tokens:
            idx = ingredient_idx[t]
            x[idx] = 1
        return x

    i = 0
    for tokens in corpus:
        A[i, :] = oh_encoder(tokens)
        i += 1

    model = TSNE(n_components = 2, learning_rate = 200)
    tsne_features = model.fit_transform(A)

    df['X'] = tsne_features[:, 0]
    df['Y'] = tsne_features[:, 1]
    return df
```

```python
df_all = pd.DataFrame()
for op_1 in option_1:
    for op_2 in option_2:
            temp = my_recommender(op_1, op_2)
            temp['Label'] = op_1 + '_' + op_2
            df_all = pd.concat([df_all, temp])

# Save the file
df_all.to_csv('cos_TSNE.csv', encoding = 'utf-8-sig', index = False)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
des_tfidf = tfidf.fit_transform(data3['ingredients'])
des_tfidf.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
des_cosine_sim = cosine_similarity(des_tfidf, des_tfidf)
des_cosine_sim.shape
```

```python
def recommend(product, sim):
    products = data3['Label'].values

    product_idx = np.where(products == product)[0]
    similarity_scores = sim[product_idx].ravel()
    products_and_scores = dict(zip(products, similarity_scores)).items()
    top_10 = sorted(products_and_scores, key= lambda x: x[1], reverse= True)[1:11]

    idx = [product_idx[0]]
    for i, (name, score) in enumerate(top_10):
        idx.append(list(np.where(products == name)[0])[0])
    return data3.iloc[idx]
```
